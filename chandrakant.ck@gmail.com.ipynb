```python
uc_merge=uc_merge.drop_duplicates()  #dropping the duplicates rows..
uc_merge
```

```python
# visually checking the Null/Nan values in the Dataset/Dataframe using heatmap

sns.heatmap(uc_merge.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
pd.DataFrame(uc_merge[['engV', 'drive']].isna().sum()) #let us check null values for engV and drive
```

### Let us first fill null values of engV and drive.

```python
uc_merge['engV'].unique()
```

```python
uc_merge['engV'].fillna(uc_merge['engV'].median(), inplace=True) # replace engV null values with median of engV
uc_merge.describe()
```

```python
uc_merge['drive'].fillna(value='front',axis=None, inplace=True)
uc_merge.describe(include='all')                       # replace drive null values with front ,as it is more than 50%
```

```python
pd.DataFrame(uc_merge[['engV', 'drive']].isna().sum()) # lets check , if still any null values in engV and drive
```

- There is one more problem with engV , that is no car in world is more than 10L(10000CC), while the column shows highest litre is 99.99 so it seems after 10, decimal is forgotton by mistake.

```python
#def engV_correction(x):
    #if x >10.0:
        #return x*0.1
    #return x
#uc_merge['engV'].apply(engV_correction)
#uc_merge
#def engV_correction(x):                                                                # We are going to apply this function
    #num = x
    #if x>10:
        #num *= 0.1
   # return num
#uc_merge['engV'].apply(engV_correction)
uc_merge['engV']= uc_merge['engV'].apply(lambda x:x/10 if x>10 else x) # dividing engV values by 10, wherever it is more than 10
uc_merge
```

```python
uc_merge['engV'].unique() 
```

- There is still a problem as few engV are less than 1.0, so lets multiply them by 10

```python
uc_merge['engV']= uc_merge['engV'].apply(lambda y:y*10 if y<1 else y) # multiply engV values by 10, wherever it is less than 1
uc_merge
```

```python
uc_merge['engV'].unique() # lets chck if all values are in range 1 -10
```

```python
uc_merge[uc_merge['price']==9200].count() # lets check how many car price values are in median price.
```

```python
uc_merge['price']=uc_merge['price'].replace(0,np.nan) #replace zero values of price with median Via null values
uc_merge['price']=uc_merge['price'].fillna(uc_merge.groupby(['car','body'])['price'].transform('median'))
uc_merge
```
